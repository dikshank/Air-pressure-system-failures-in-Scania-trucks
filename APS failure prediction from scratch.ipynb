# READ THE DATA

```python
import pandas as pd
import numpy as np

train_data = pd.read_csv("aps_failure_training_set.csv", skiprows=20)

test_data = pd.read_csv("aps_failure_test_set.csv",skiprows=20)
```

```python
data = pd.concat([train_data,test_data])
```

# Replacing "na" values with "np.nan" values

```python
data.replace(to_replace="na", value = np.nan,inplace=True)
```

```python
data
```

# dropping columns which had more NaN values than decided threshold

```python
data.dropna(axis=1,inplace=True,thresh=(0.8*data.shape[0]))
```

```python
data
```

# Filling the rest of missing values using __*SimpleImputer*__ from sklearn

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer()
```

```python
data_array = imputer.fit_transform(X=data.iloc[:,1:])
```

```python
data_array
```

```python
data_columns = data.columns
data_labels = data['class']
```

```python
data = pd.DataFrame(data=data_array,columns=data_columns[1:])
```

```python
data
```

# Now let's do the quantization of column values using __*pd.qcut*__ 

we can do the binning of these kind of values (quantiles are bins)

```python
unique_values = dict()
for column in data.columns:
    unique_values[column] = len(data[column].unique())
```

```python
# unique_values
```

```python
data.drop(['cd_000'],axis=1,inplace=True)     # as cd_000 has only single value so we are gonna drop it
```

```python
for column in data.columns[0:145]:
    data[column] = pd.qcut(x=data[column],q=5,duplicates='drop').cat.codes
```

```python
data
```

# Now doing the __*one-hot encoding*__ of column values

```python
np_array_list = list()
for column in data.columns[0:data.shape[1]-1]:
    np_array_list.append(np.eye(10,10)[data[column]])
data_array = np.concatenate(np_array_list,axis=1)
```

```python
data_array
```

```python
data_array.shape
```

# let's apply dimensionality reduction

```python
from sklearn.decomposition import PCA
```

```python
pca_obj = PCA(n_components=10)
```

```python
reduced_data_array = pca_obj.fit_transform(data_array)
```

```python
reduced_data_array.shape
```

```python
data_array
```

```python
reduced_data_array
```

# Let's do data splitting

```python
train_data_len = int(0.8*train_data.shape[0])
train_data_len
```

```python
cv_data_len = int(0.2*train_data.shape[0])
cv_data_len
```

```python
test_data_len = test_data.shape[0]
test_data_len
```

```python
train_data_array = reduced_data_array[0:train_data_len,:]

cv_data_array = reduced_data_array[train_data_len:train_data_len+cv_data_len,:]

test_data_array = reduced_data_array[train_data_len+cv_data_len:,:]
```

```python
train_data_array.shape
```

```python
cv_data_array.shape
```

```python
test_data_array.shape
```

```python
train_data_array
```

# Let's apply smote on training dataset

```python
train_data_new = train_data.iloc[0:train_data_len,:]
```

```python
train_data_new.shape
```

```python
train_data['class']
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote_obj = SMOTE(sampling_strategy='minority',k_neighbors=5)
```

```python
X_oversampled, Y_oversampled = smote_obj.fit_resample(X=train_data_array,y=train_data_new['class'])
```

```python
X_oversampled.shape
```

```python
train_data_new = pd.DataFrame(X_oversampled)
train_data_new['class'] = Y_oversampled
```

```python
train_data_new
```

```python
cv_data_new = pd.DataFrame(cv_data_array)
```

```python
np.array(data_labels.iloc[train_data_len:train_data_len+cv_data_len])
```

```python
cv_data_new['class'] = np.array(data_labels.iloc[train_data_len:train_data_len+cv_data_len])
```

```python
cv_data_new
```

# Applying Gaussian Naive Bayes algorithm on training data 

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gaussian_nb = GaussianNB()
```

```python
train_data_new.shape
```

```python
gaussian_nb.fit(X=np.array(train_data_new.iloc[:,0:train_data_new.shape[1]-1]),y=np.array(train_data_new['class']))
```

# Finally predicting the labels using CV data and calculating the accuracy

```python
predicted_labels = gaussian_nb.predict(X=np.array(cv_data_new.iloc[:,0:cv_data_new.shape[1]-1]))
```

```python
predicted_labels
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_true=np.array(cv_data_new['class']),y_pred=predicted_labels))
```
